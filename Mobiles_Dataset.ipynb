```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Mobiles.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.subplots(figsize=(10,6))
sns.heatmap(df.isnull(), cmap='viridis')
plt.show()
```

# **Observation**
*   Memory, Storage and Rating are the three columns with some of the null blocks.
*   Rating has maximum number (144) of null block.

```python
df["Rating"].replace(np.nan, df["Rating"].mean(), inplace=True)
df.drop('Memory', axis=1, inplace=True)
df.drop('Storage', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
df = df.sort_values("Brand")
```

```python
df.Brand.value_counts()
```

```python
brand_top = df.Brand.value_counts().head(3)
brand_bottom = df.Brand.value_counts().tail(3)
brand_names1 = brand_top.index
brand_counts1 = brand_top.values
brand_names2 = brand_bottom.index
brand_counts2 = brand_bottom.values
```

```python
plt.subplots(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.pie(brand_counts1, labels=brand_names1, autopct='%1.1f%%')
plt.title('Top 3 available Brands')

plt.subplot(1, 2, 2)
plt.pie(brand_counts2, labels=brand_names2, autopct='%1.1f%%')
plt.title('Bottom 3 available Brands')

plt.show()
```

# **Observation**
*   Samsung has the most number of mobile phones followed by Apple and realme.
*   IQOO has the least number of mobile in the market.

```python
groups_brand = df.groupby('Brand')
groups_brand['Selling Price'].mean()
```

```python
plt.subplots(figsize=(10,6))
sns.boxplot(x=df["Brand"], y=df["Selling Price"] ,palette="bright")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(15, 5))

plt.subplot(1, 2, 1)
most_costly_model = df.loc[df.groupby("Brand")["Selling Price"].idxmax()]
sns.barplot(x="Model",y="Selling Price",data=most_costly_model,palette="flare")
plt.title('Most Expensive Model')
plt.xticks(rotation = 90)

plt.subplot(1, 2, 2)
least_costly_model = df.loc[df.groupby("Brand")["Selling Price"].idxmin()]
sns.barplot(x="Model",y="Selling Price",data=least_costly_model,palette="flare")
plt.title('Most Cheapest Model')
plt.xticks(rotation = 90)

plt.show()
```

# **Observation**
*   Apple is the most costly selling brand followed by Samsung.
*   Google makes more mid to premium range mobile than Samsung .
*   iPhone 13 and 13 pro max (of Apple) and Fold 2 5G (of Samsung) are the most expensive models in the market .
*   GT 1200 R/I/M (of Samsung) and L700 (of Gionee) and Ta - 1010/105 (of Nokia) are the least expensive models in the market .

```python
plt.subplots(figsize=(10,6))
sns.lineplot(x=df["Brand"],y=df["Selling Price"], marker="o", ci=None)
plt.xticks(rotation=90)
plt.title("Average Selling Price by Brand")
plt.grid()
plt.show()
```

# **Observation**
*   Apple Brand has the highest average selling price among all followed by Google Pixel then followed by IQOO.
*   Gionee Brand has the least average selling price among all.

```python
plt.subplots(figsize=(10,6))
df["Price Range"] = pd.cut(df["Selling Price"], bins=[0, 15000, 40000, float("inf")], labels=["Low Range", "Mid Range", "Premium Range"])
sns.countplot(x="Price Range",data=df,palette="flare")
plt.title("Count of Phones in Different Price Ranges")
plt.show()
```

# **Observation**
*   Most Smartphones Belong to Low Range Category (0-15000 Rs).

```python
common_colors = df["Color"].value_counts().head(10)
plt.figure(figsize=(8, 8))
plt.pie(common_colors,labels=common_colors.index,autopct='%1.1f%%')
plt.title("Most Common Colors Offered")
plt.show()
```

# **Observation**
*   Most Common Color offered is Black followed by Gold then followed by White and then followed by Blue.

```python
plt.subplots(figsize=(10,6))
sns.boxplot(x="Brand",y="Rating",data=df,palette="bright")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(15, 5))

plt.subplot(1, 2, 1)
most_rated_model = df.loc[df.groupby("Brand")["Rating"].idxmax()]
sns.lineplot(x="Model",y="Rating",data=most_rated_model, marker="o")
plt.title('Highest Rated Model')
plt.xticks(rotation = 90)

plt.subplot(1, 2, 2)
most_rated_model = df.loc[df.groupby("Brand")["Rating"].idxmin()]
sns.lineplot(x="Model",y="Rating",data=most_rated_model, marker="o")
plt.title('Lowest Rated Model')
plt.xticks(rotation = 90)
plt.show()
```

# **Observation**
*   Apple is the most rated brand followed by Google Pixel.
*   iPhone 7 Plus (of Apple), SM-B310EZDDINS (of Samsung), X60 pro (of Vivo) are rated 5 star.
*   Redmi Note 8 (of Xiaomi) and Quench X T3 (of Motorola) are the two worst rated mobiles.

```python
df['Discount']=df['Original Price']-df['Selling Price']
df['Discount in %']=(df['Discount']/df['Original Price'])*100
df.head()
```

```python
df1=df[df['Original Price']<df['Selling Price']]
df1
```

```python
df1=df[~(df['Original Price']<df['Selling Price'])]
df1
```

```python
df2=df1[(df1['Original Price']==df1['Selling Price'])]
df2
```

```python
df2=df1[~(df1['Original Price']==df1['Selling Price'])]
df2
```

```python
plt.subplots(figsize=(10,6))
sns.barplot(x=df2["Brand"],y=df2["Discount"], palette="flare", ci=None)
plt.title('Average Discount')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(18,5))
sns.boxplot(x=df2["Brand"], y=df2["Discount"], palette="bright")
plt.show()
```

```python
plt.subplots(figsize=(15, 5))

plt.subplot(1, 2, 1)
most_discounted_model = df2.loc[df2.groupby("Brand")["Discount"].idxmax()]
sns.lineplot(x="Model",y="Discount",data=most_discounted_model, marker="o")
plt.title('Highest Discounted Model')
plt.xticks(rotation = 90)

plt.subplot(1, 2, 2)
least_discounted_model = df2.loc[df2.groupby("Brand")["Discount"].idxmin()]
sns.lineplot(x="Model",y="Discount",data=least_discounted_model, marker="o")
plt.title('Lowest Discounted Model')
plt.xticks(rotation = 90)
plt.show()
```

# **Observation**
*   Google Pixel gives the most average discount followed by LG.
*   Razr (of Motorola) and ROG (of ASUS) are highest discounted model.
